```python
# IMPORT STATEMENTS
# (NONE)

# FUNCTION DEFINITIONS
# 1. Create the function is_palindrome with a single parameter: word.
def is_palindrome(word):
#    -Take the word given and make it case insensitive.
  word = word.upper()
#    -Reverse the word and assign it to a variable.
  inverse_word = word[::-1]
#    -Create an if that compares the word and the reversed word variable.
  if word == inverse_word:
#      -If they are the same, return true, false if not.
    return True
  return False

# PROGRAM LOGIC
# Ask the user to input the word they want to check.
word = input( 'Type a word: ' )
# Call the function with the user's input.
is_palindrome(word)
# Get the returned value of the function in an if.
if is_palindrome(word):
# Print This word is a palindrome! if the value returned is true.
  print( 'This word is a palindrome!' )
# Else print This word is not a palindrome. if false.
else:
  print( 'This word is not a palindrome.' )
```
